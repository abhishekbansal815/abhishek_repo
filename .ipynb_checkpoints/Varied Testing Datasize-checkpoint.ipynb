```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot For Accuracy based on Unigram CTF vs Testing Datasize
data = {'Testing Datasize': [20, 40, 60, 80, 100],
       'Accuracy': [0.80, 0.875, 0.8834, 0.9, 0.93]}
df1 = pd.DataFrame(data, columns = ['Testing Datasize', 'Accuracy'])
df1.plot(x = 'Testing Datasize', y = 'Accuracy', kind = 'line', title = 'Accuracy Score based on Unigram CTF vs Testing Datasize')
plt.show()
```
